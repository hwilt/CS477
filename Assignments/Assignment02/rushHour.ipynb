### Neighboring States (10 Points)

```python
s = State()
s.load_puzzle("jams/veryeasy.txt")
neighbors = s.get_neighbors()
# Plot everything
N = len(neighbors)+1
K = int(np.ceil(np.sqrt(N))) # Figure out a square grid to put these on

plt.figure(figsize=(10, 10))
plt.subplot(K, K, 1)
plt.title("Original")
s.plot()
for i, n in enumerate(neighbors):
    plt.subplot(K, K, i+2)
    n.plot()
    plt.title("Neighbor {}".format(i))
plt.tight_layout()

print("# of neighbors: {}".format(len(neighbors)))
```

### Breadth-First Tree Search (15 Points)

**VERY** Easy Puzzle

```python
s = State()
s.load_puzzle("jams/veryeasy.txt")

tic = time.time() # Time the solution
solution = s.solve()
toc = time.time()
print("Elapsed time: {:.3f} seconds".format(toc-tic))
for i, s in enumerate(solution):
    plt.figure()
    s.plot()
    plt.title("Step {}".format(i))
print(len(solution))
```
